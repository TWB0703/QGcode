```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.sans-serif'] = ['SimHei']

train_df=pd.read_csv('./data/train.csv')
test_df=pd.read_csv('./data/test.csv')

train_df['Age'].fillna(train_df['Age'].mean(),inplace=True)
train_df['Embarked'].fillna(train_df['Embarked'].mode(),inplace=True)

plt.figure(figsize=(16,10))
plt.subplot(2,3,1)
sns.barplot(x='Sex',y='Survived',data=train_df,palette='Set2')
plt.title('性别与生还率')
plt.xlabel('性别')
plt.ylabel('生还率')
plt.xticks(ticks=[0,1],labels=['男','女'])

plt.subplot(2,3,2)
sns.barplot(x='Pclass',y='Survived',data=train_df,palette='Set1')
plt.title('舱位与生还率')
plt.xlabel('舱位等级')
plt.ylabel('生还率')

plt.subplot(2,3,3)
sns.barplot(x='Embarked',y='Survived',data=train_df,palette='Set2')
plt.title('登船港口与生还率')
plt.xlabel('登船港口')
plt.ylabel('生还率')

plt.subplot(2,3,4)
sns.histplot(train_df[train_df['Survived']==1]['Age'],bins=20,color='green',kde=True,label='生还')
sns.histplot(train_df[train_df['Survived']==0]['Age'],bins=20,color='red',kde=True,label='未生还')
plt.legend()
plt.title('年龄分布与生还情况')
plt.xlabel('年龄')
plt.ylabel('人数')

train_df['FamilySize'] = train_df['SibSp'] + train_df['Parch'] + 1
plt.subplot(2, 3, 5)
sns.barplot(x='FamilySize', y='Survived', data=train_df, palette='coolwarm')
plt.title('家庭规模与生还率')
plt.xlabel('家庭规模')
plt.ylabel('生还率')

# 图6：票价与生还（去除异常值）
plt.subplot(2, 3, 6)
sns.boxplot(x='Survived', y='Fare', data=train_df, showfliers=False)
plt.title('票价与生还情况')
plt.xlabel('是否生还')
plt.ylabel('票价')

plt.tight_layout()
plt.show()






```
